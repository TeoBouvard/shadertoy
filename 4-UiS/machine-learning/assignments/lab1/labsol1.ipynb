# 1D probability density function

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from pdffuns import norm1D, norm2D
```

```python
plt.rcParams['figure.figsize'] = [10, 8]
```

```python
x = np.arange(-10, 10.5, 0.5).reshape(-1, 1)
mu = 1
sigma = 2
p = norm1D(mu, sigma, x)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, x*0, 'g.')
ax.plot(x, p, 'b')
ax.plot(x, p, 'r.')
ax.grid(color='gray', linestyle='--')
plt.show()
```

# 2D probability density function

```python
x1 = np.arange(-10, 10.5, 0.5).reshape(-1, 1)
x2 = np.arange(-9, 10.5, 0.5).reshape(-1, 1)
grid = np.meshgrid(x1, x2)
mu = np.array([1, 1]).reshape(2, 1)
covariance_matrix = np.array([5, 3, 3, 5]).reshape(2, 2)
p = norm2D(mu, covariance_matrix, grid)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
#ax.scatter(*grid, c='green')
ax.plot_surface(*grid, p)
```
