```python
import pandas as pd
from pandas_datareader.data import DataReader
from matplotlib import pyplot as plt
%matplotlib inline
```

## Part 1

It is not specified what a `period` is. I assume it is a start date and an end date as strings.

```python
def show_correlation(stock_name, start, end):
    stock = DataReader(stock_name, start=start, end=end, data_source='yahoo')[['Open', 'Close']]
    business_month_end = pd.date_range(start, end, freq='BM')
    month_end = stock[stock.index.isin(business_month_end)]
    plt.scatter(x=month_end.index, y=month_end['Open'])
    plt.scatter(x=month_end.index, y=month_end['Close'])
    plt.legend(['Open', 'Close'])
    plt.xlabel("Date")
    plt.ylabel("Price in $")
    plt.title("Monthly correlation of open and close prices of {} stock between {} and {}".format(stock_name, start, end))
```

```python
show_correlation('AAPL', '2003-01-03', '2018-01-01')
```
